```python
from matplotlib import pyplot as plt
```

```python
# Line Graph

x = [5,7,10,15,25,30]
y = [6,10,15,20,30,33]
plt.figure(figsize=(6,5))   # Adjusting the size of the graph
plt.plot(x,y, marker='o', color='red')
plt.xlabel("Sales")
plt.ylabel("Marketing Expenses")
plt.title("Sales Report")
plt.show()
```

```python
# Multiline line graph

import numpy as np
from matplotlib import pyplot as plt

y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])

plt.plot(y1, marker = 'o')
plt.plot(y2,marker = 'o', linstyle="dashed")

plt.show()
```

```python
# Bar chart

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]

plt.figure(figsize=(4,3))
plt.bar(langs, students,color='#d12f24',width=0.8)

plt.title('Students Enrolled')
plt.ylabel('Students')
plt.xlabel('Courses')
plt.show()
```

```python
# Horizontal Bar chart

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]

#plt.figure(figsize=(4,3))
plt.barh(langs,students,color='yellow', width=0.8)

plt.title('Students Enrolled')
plt.ylabel('Students')
plt.xlabel('Courses')
plt.show()
```
